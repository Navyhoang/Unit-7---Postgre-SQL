```python
import psycopg2
from config import username, password
from sqlalchemy import create_engine
import pandas as pd
import matplotlib.pyplot as plt

# Create engine and connect to PostgreSQL database
engine = create_engine(f'postgresql://{username}:{password}@localhost:5432/SQl Homework')

connection = engine.connect()
```

```python
# Read csvs from database using PANDAS
salary_df = pd.read_sql("SELECT * FROM salaries", connection)
salary_df.head()
```

```python
employess_df = pd.read_sql("SELECT * FROM employees", connection)
employess_df.dropna(how='any')
employess_df.head()
```

```python
# Merge 2 tables above
merge_df = pd.merge(salary_df,employess_df, on="emp_no", how='left')
merge_df.dropna(how='any')
merge_df.head()
```

# Create a histogram using Matplotlib

```python
plt.figure(figsize=(10,6))
plt.hist(merge_df['salary'], align='mid', color='pink')
plt.xlabel('Salary')
plt.ylabel('Count')
plt.xlim(35000,125000)
plt.grid()
plt.title('Employees Salary Histogram')
plt.show()
```

```python
# The most common salary range is between $40,000 - $50,000, followed by $50,000-$75,000.
# Less people are earning a salary above $80,000
```

# Create a bar charge of average salary by title

```python
titles_df = pd.read_sql("SELECT * FROM titles", connection)
titles_df
```

```python
final_df = pd.merge(merge_df,titles_df, left_on="emp_title", right_on='title_id', how='outer')
final_df
```

```python
# Get average salary by title using pandas
avg_salary_by_title = round(final_df.groupby('title')['salary'].mean(),2)

avg_salary_by_title_df = pd.DataFrame({
                      
                         "Average salary": avg_salary_by_title
                         
                         })

avg_salary_by_title_df
```

```python
# Create a bar chart using Matplotlib
title = avg_salary_by_title_df.index

plt.figure(figsize=(10,6))
plt.bar(avg_salary_by_title_df.index, avg_salary_by_title_df['Average salary'], color ='orange', align='center')
plt.title('Average Salary by Employee Title')
plt.xlabel('Average salary ($)')
plt.xticks(title, rotation =90)
plt.ylabel('Employee title')
plt.ylim(48000,60000)
plt.grid()
plt.show()
```

```python
# Average salary is highest for employee titles s0001 and s0002
```
